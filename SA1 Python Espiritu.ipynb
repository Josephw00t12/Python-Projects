```python
#F Answer
from collections import defaultdict

letter_counts = defaultdict(int)
with open('wordlist.txt', 'r') as f:
    for word in f:
        for letter in word.strip():
            letter_counts[letter] += 1

for letter, count in sorted(letter_counts.items()):
    print(f"{letter}: {count}")
```

```python
#G Word Guessing Game
import random

def word_selection():
    with open('wordlist.txt', 'r') as f:
        words = f.readlines()
        word_list = list(random.choice(words).strip())
    return word_list

def display_logic(word, guessed):
    display = []
    for letters in word:
        if letters in guessed:
            display.append(" " + letters + " ")
        else:
            display.append('___')
    return display

def display_words(word_choice, letters_guessed):
    print(f"The word is: {len(word_choice)} letters long")
    for letters in display_logic(word_choice, letters_guessed):
        print(letters, end=" ")

def user_guess_letter():
    while True:
        guess = input("Enter a Letter: ").lower()
        if guess.isalpha() and len(guess) == 1:
            print()
            return guess
        elif len(guess) > 1 or len(guess) <= 0:
            print("Invalid Amount")
        else:
            print("Not a Letter!")

def game():
    word = word_selection()
    correct_guessed_letters = []
    incorrect_guessed_letters = []
    max_guess = 10
    
    while max_guess > 0:
        display_words(word, correct_guessed_letters)
        
        print(f"\nYou have {max_guess} guesses left\n\n")
        
        guess = user_guess_letter()
        
        if guess in correct_guessed_letters or guess in incorrect_guessed_letters:
            print(f"You already have guessed letter '{guess}' try another one")
        elif guess in word:
            print(f"Correct there is a '{guess}' in the word")
            correct_guessed_letters.append(guess)
        else:
            incorrect_guessed_letters.append(guess)
            max_guess -=1
            print(f"Incorrect there is no '{guess}' in the word")
            
        if all(letter in correct_guessed_letters for letter in word):
            word_win = ''.join(word)
            print()
            print(f"{max_guess} guesses left, Yay! you won the game")
            print(f"The word was {word_win.capitalize()}")
            break
        
    if max_guess <= 0:        
        print(f"{max_guess} guesses left, You lost the game")
        print(f"The word was {word_win.capitalize()}")
    
game()
```

```python
#Question 2
import time

def fib_iter(num):
    if num < 0:
        return 0
    elif num == 0:
        return 0
    elif num == 1:
        return 1
    else:
        a,b = 0,1
        for i in range(2,num+1):
            a,b = b, a+b
        return b
    
ni = 10000

start_i = time.time()
print(fib_iter(ni))
end_i = time.time()

print(f"Time taken to compute {ni} Fibonacci numbers iteratively: {end_i-start_i} seconds")
```

```python
def fib_recur(num):
    if num < 0:
        return 0
    elif num == 0:
        return 0
    elif num == 1:
        return 1
    else:
        return fib_recur(num-1) + fib_recur(num-2)
    
nr = 38

start_r = time.time()
print(fib_recur(nr))
end_r = time.time()

print(f"Time taken to compute {nr} Fibonacci numbers recursively: {end_r-start_r} seconds")
```
