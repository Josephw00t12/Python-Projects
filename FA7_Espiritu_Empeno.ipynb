```python
import math as m

#Classes of Circle and Point
class Circle:
    def __init__(self, center, radius):
        self.center = center
        self.radius = radius

class Point:
    def __init__(self, point):
        self.x, self.y = point

        
# Using Pythagorean Theorem to find Distance of two points and
# Compare the Distance to the Radius True/False
# Delta x and y
def point_in_circle(_circle, _point):
    radicand = ((_circle.center.x - _point.x)**2) + ((_circle.center.y - _point.y)**2)
    distance = round(m.sqrt(radicand), 3)
    return distance <= _circle.radius, distance
    
# Make a Point function
def pointMaker():
    while True:
        try:
            x = int(input("X-Value: "))
            y = int(input("Y-Value: "))
            break
        except ValueError:
            print("\nInvalid Number")
    return x,y


# Question 1
# Instantiate
mainCircle = Circle(Point((150,100)),75)

print(f"({mainCircle.center.x},{mainCircle.center.y}) Center Point of the Circle")
print(f"{mainCircle.radius} The Radius of a Circle")

madePoint = Point(pointMaker())

print(f"({madePoint.x},{madePoint.y}) A Point")

ans, d = point_in_circle(mainCircle, madePoint)
print(f"The Radius of the Circle is {mainCircle.radius}")
print(f"The Answer is: {ans} because of the distance of {d}")
print("Because of the compariing if the distance is less than or equal to the radius of the circle which is 75")
```

```python
#Question 2

def distance_of_two_points(point1, point2):
    radicand = ((point1.x - point2.x)**2) + ((point1.y - point2.y)**2)
    distance = round(m.sqrt(radicand))
    return distance

def makeRectangle():
     rect = []
     print("To make a rectangle you need 4 points")
     for i in range(4):
         if i+1 == 1:
             print("Rectangle's Top Left Point")
             rect.append(Point(pointMaker()))
         elif i+1 == 2:
             print("Rectangle's Top Right Point")
             rect.append(Point(pointMaker()))
         elif i+1 == 3:
             print("Rectangle's Bottom Left Point")
             rect.append(Point(pointMaker()))
         elif i+1 == 4:
             print("Rectangle's Bottom Right Point")
             rect.append(Point(pointMaker()))
     return rect

def confirmRectangle(rect):
    # 1. if Top Left Top Right Distance == Bottom Left Bottom Right Distance
    # 2. Bottom and Left Distance Squared is Hypotenuse Squared
    topDistance = distance_of_two_points(rect[0],rect[1])
    bottomDistance = distance_of_two_points(rect[2],rect[3])
    leftDistance = distance_of_two_points(rect[0],rect[2])
    rightDistance = distance_of_two_points(rect[1],rect[3])

    if topDistance == bottomDistance and leftDistance == rightDistance:
        return True
    return False

def rect_in_circle(circle,rect):
    ans1, d1 = point_in_circle(circle, rect[0])
    ans2, d2 = point_in_circle(circle, rect[1])
    ans3, d3 = point_in_circle(circle, rect[2])
    ans4, d4 = point_in_circle(circle, rect[3])
    
    print(f"Distance1= {d1}  Distance2= {d2}  Distance3= {d3}  Distance4= {d4}")
    if ans1 and ans2 and ans3 and ans4:
        return True
    return False

# Rectangles within Circle:
# TopLeft(90,140) TopRight(200,140) BottomLeft(90,80) BottomRight(200,80)

# Rectangles not within Circles
# TopLeft(95,100) TopRight(220,100) BottomLeft(95,60) BottomRight(220,60)

# Not Rectangles
# TopLeft(100,100) TopRight(220,100) BottomLeft(95,60) BottomRight(220,60)

# Rectangle with a Single Point on the Circle
# TopLeft(50,78) TopRight(79,78) BottomLeft(50,70) BottomRight(79,70)

# Rectangle with no Points on the Circle but Overlaps
# TopLeft(50,100) TopRight(250,100) BottomLeft(50,50) BottomRight(250,50)

madeRectangle = makeRectangle()
is_rectangle = confirmRectangle(madeRectangle)

print()
if not is_rectangle :
    print("The Four Points Given is not a Rectangle")
elif is_rectangle and rect_in_circle(mainCircle, madeRectangle):
    print("The Rectangle is within or on the Boundaries of the Circle given")
else: 
    print("The Rectangle is not within or on the Boundaries of the Circle given")
```

```python
# Question 3
def rect_circle_overlap(circle,rect):
    ans1, d1 = point_in_circle(circle, rect[0])
    ans2, d2 = point_in_circle(circle, rect[1])
    ans3, d3 = point_in_circle(circle, rect[2])
    ans4, d4 = point_in_circle(circle, rect[3])
    
    # Use Points as Basis for Overlapping
    print(f"Distance1= {d1}  Distance2= {d2}  Distance3= {d3}  Distance4= {d4}")
    if ans1 or ans2 or ans3 or ans4:
        return True
    # Line Intersect Basis
    else:
        xrect1, yrect1 = rect[0].x, rect[0].y
        xrect2, yrect2 = rect[1].x, rect[1].y
        xrect3, yrect3 = rect[2].x, rect[2].y
        xrect4, yrect4 = rect[3].x, rect[3].y

        # Top line
        distx1 = abs(circle.center.x - (xrect1 + xrect2)/2)
        disty1 = abs(circle.center.y - (yrect1 + yrect2)/2)
        if distx1 > (xrect2 - xrect1)/2 + circle.radius:
            intersects1 = False
        elif disty1 > (yrect2 - yrect1)/2 + circle.radius:
            intersects1 = False
        elif distx1 <= (xrect2 - xrect1)/2:
            intersects1 = True
        elif disty1 <= (yrect2 - yrect1)/2:
            intersects1 = True
        else:
            dx1 = distx1 - (xrect2 - xrect1)/2
            dy1 = disty1 - (yrect2 - yrect1)/2
            if dx1**2 + dy1**2 <= circle.radius**2:
                intersects1 = True
            else:
                intersects1 = False
        # Bottom Line
        distx2 = abs(circle.center.x - (xrect3 + xrect4)/2)
        disty2 = abs(circle.center.y - (yrect3 + yrect4)/2)
        if distx2 > (xrect4 - xrect3)/2 + circle.radius:
            intersects2 = False
        elif disty2 > (yrect4 - yrect3)/2 + circle.radius:
            intersects2 = False
        elif distx2 <= (xrect4 - xrect3)/2:
            intersects2 = True
        elif disty2 <= (yrect4 - yrect3)/2:
            intersects2 = True
        else:
            dx2 = distx2 - (xrect4 - xrect3)/2
            dy2 = disty2 - (yrect4 - yrect3)/2
            if dx2**2 + dy2**2 <= circle.radius**2:
                intersects2 = True
            else:
                intersects2 = False

        # Left Line
        distx3 = abs(circle.center.x - (xrect3 + xrect1)/2)
        disty3 = abs(circle.center.y - (yrect3 + yrect1)/2)
        if distx3 > (xrect1 - xrect3)/2 + circle.radius:
            intersects3 = False
        elif disty3 > (yrect1 - yrect3)/2 + circle.radius:
            intersects3 = False
        elif distx3 <= (xrect1 - xrect3)/2:
            intersects3 = True
        elif disty3 <= (yrect1 - yrect3)/2:
            intersects3 = True
        else:
            dx3 = distx3 - (xrect1 - xrect3)/2
            dy3 = disty3 - (yrect1 - yrect3)/2
            if dx3**2 + dy3**2 <= circle.radius**2:
                intersects3 = True
            else:
                intersects3 = False

        # Right Line
        distx4 = abs(circle.center.x - (xrect2 + xrect4)/2)
        disty4 = abs(circle.center.y - (yrect2 + yrect4)/2)
        if distx4 > (xrect2 - xrect4)/2 + circle.radius:
            intersects4 = False
        elif disty4 > (yrect2 - yrect4)/2 + circle.radius:
            intersects4 = False
        elif distx4 <= (xrect2 - xrect4)/2:
            intersects4 = True
        elif disty4 <= (yrect2 - yrect4)/2:
            intersects4 = True
        else:
            dx4 = distx4 - (xrect2 - xrect4)/2
            dy4 = disty4 - (yrect2 - yrect4)/2
            if dx4**2 + dy4**2 <= circle.radius**2:
                intersects4 = True
            else:
                intersects4 = False
                
        # Check if any are true
        if intersects1 or intersects2 or intersects3 or intersect4:
            return True
    return False

print(rect_circle_overlap(mainCircle, madeRectangle))
```
