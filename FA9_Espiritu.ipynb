```python
#Fill NA in likes with 0, NA in comments with 0, and drop others.
dropFill = pd.DataFrame()

dropFill['Likes'] = data['Likes'].fillna(0)
dropFill['Comments'] = data['Comments'].fillna(0)

# drop all other columns
amount = dropFill[['Likes', 'Comments']].isnull().sum()

print(dropFill)
amount
```

```python
#Describe (descriptive measures: count, mean, std, min, 25%, 50%, 75%, max) the cleaned data.
data['Likes'].fillna(0)
data['Comments'].fillna(0)
data.describe().T
```

```python
#Explore album type and display counts for album, single, and compilation.

dcount = data['Album_type'].value_counts()
print(dcount)
```

```python
#Create a pie chart for Album Types as a Title and make some remarks.
pie = dcount.plot.pie(title='Album Types', figsize=(4,4))
# Majority of the songs are in an album release and singles follows next then compilations as the least used type
```

```python
#Analyze songs by artist and views-youtube, stream-spotify.
grouped = data.groupby('Artist').agg({'Views': 'sum', 'Stream': 'sum'})

#Top 15 Artists by order in views and streams
top = grouped.sort_values(['Views','Stream'], ascending=False).head(15)
print(f"Top 15 \n{top}")
#Bottom 15 Artists by order in views and streams
bottom = grouped.sort_values(['Views','Stream'], ascending=True).head(15)
print(f"\n\nBottom 15 \n{bottom}")
```

```python
#Create bar charts for Top 10 Artists on Youtube and Top 10 Artists on Spotify in 1 row of display.

you_sorted = grouped.sort_values(['Views'], ascending=False).head(10)
spot_sorted = grouped.sort_values(['Stream'], ascending=False).head(10)

# Change names into numbers and rank them
ranked_you = pd.DataFrame({'Rank': range(1, len(you_sorted)+1),'Views Youtube': you_sorted['Views']}, index= you_sorted.index)
ranked_spot = pd.DataFrame({'Rank': range(1, len(spot_sorted)+1),'Streams Spotify': spot_sorted['Stream']}, index= spot_sorted.index)

label, bars = mpl.pyplot.subplots(1,2, figsize=(15,6))

#Bars Youtube Top 10 Artsits change to names
bars[0].bar(ranked_you['Rank'], ranked_you['Views Youtube'])
bars[0].set_xticks(range(1, len(ranked_you) + 1))
bars[0].set_xticklabels(you_sorted.index, rotation=90)
bars[0].set_title('Top 10 Artists on Youtube')
bars[0].set_xlabel('Artist')
bars[0].set_ylabel('Views')

#Bars Spotify Top 10 Artsits change to names
bars[1].bar(ranked_spot['Rank'], ranked_spot['Streams Spotify'])
bars[1].set_xticks(range(1, len(ranked_spot) + 1))
bars[1].set_xticklabels(spot_sorted.index, rotation=90)
bars[1].set_title('Top 10 Artists on Spotify')
bars[1].set_xlabel('Artist')
bars[1].set_ylabel('Streams')

label.suptitle('Top 10 Artists on Youtube and Spotify')

# Make some remarks about the Top 10 Artists on Youtube and Top 10 Artists on Spotify.

print(ranked_you)
print(ranked_spot)

# A baby oriented channel is topping Youtube which is obvious
# Albums and Artists who dont have music videos or dont usually have music videos in their music are 
# likely to be more popular in Spotify
# The views and streams are almost on par with each other
```

```python
#Analyze Top 10 songs on the basis of views, comments, and likes: with titles Top 10 Songs with Most Views,
#Top 10 Songs with Most Comments, and Top 10 Songs with Most Likes.

grouped2 = data.groupby('Artist').agg({'Views': 'sum', 'Comments': 'sum', 'Likes': 'sum'})

likeSort = grouped2.sort_values(['Likes'], ascending=False).head(10)
commentSort = grouped2.sort_values(['Comments'], ascending=False).head(10)
viewSort = grouped2.sort_values(['Views'], ascending=False).head(10)


rankedViews = pd.DataFrame({'Rank': range(1, len(viewSort)+1),'Song Views': viewSort['Views']}, index=viewSort.index)
rankedLikes = pd.DataFrame({'Rank': range(1, len(likeSort)+1),'Song Likes': likeSort['Likes']}, index=likeSort.index)
rankedComments = pd.DataFrame({'Rank': range(1, len(commentSort)+1),'Song Comments': commentSort['Comments']}, index=commentSort.index)

print(rankedViews)
print(rankedLikes)
print(rankedComments)


# Make some remarks about the Top 10 Songs with Most Views, Top 10 Songs with Most Comments, and Top 10 Songs with Most Likes.

# Views have the most amount and it's most likely because of it counting people who dont like and comment the song
# Suprisingly BTS and Blackpinks which are koreans are the ones that has the most likes and Comments maybe 
# due to more avid and dedicated fanbases that leads them to engage and talk with other fans and in general like the song
# Comments are the lowest amount meaning that people dont usually comment because it has the most effort to do in a song
```

```python
#Analyzing the songs on the basis of Energy and Danceability and display a table of it.

grouped3 = data.groupby('Track').agg({'Energy': 'mean', 'Danceability': 'mean'})

sortEnergy = grouped3.sort_values(['Energy'], ascending=False).head(20).drop('Danceability', axis=1)
sortDance =  grouped3.sort_values(['Danceability'], ascending=False).head(20).drop('Energy', axis=1)

print(sortEnergy)
print(sortDance)

# Energy is the intensity and activeness of Songs (energetic tracks feel fast, loud, and noisy)
# so the songs with the rating of 1 means its high and 0 means low
# Danceability is the suitability for dancing of the flow or feeling to dance (tempo, rhythm stability, beat strength)
# same rating 1 means high and 0 means low
```

```python
# Display a bar chart of the 
# Top 10 Songs by Stream with their Energy, Danceability, and Acousticness and make some remarks about it.

top10 = data[['Track', 'Energy', 'Danceability', 'Acousticness', 'Stream']].sort_values('Stream', ascending=False).head(10)

top10 = top10.melt(id_vars='Track', var_name='Attribute', value_name='Value')

mpl.pyplot.figure(figsize=(12, 8))
sb.barplot(data=top10, x='Track', y='Value', hue='Attribute' )
mpl.pyplot.title('Top 10 Songs by Stream with Energy, Danceability, and Acousticness')
mpl.pyplot.xlabel('Song')
mpl.pyplot.ylabel('Value')
mpl.pyplot.yscale('log')
mpl.pyplot.xticks(rotation=90)

mpl.pyplot.show()

# Blinding Lights has the most streams, energy, and danceability but lowest acoustics suffers the ranks
# The highest acoustics listed is rockstar feat 21 savage but their energy and danceability did not take a hit
# Streams are way ahead in statistics and dominates the table
```

```python
# Determine the top 10 artists on the Youtube platform whose songs have more likes and 
# display a bar chart with a title Number of Likes per Artist and make some remarks.

grouped5 = data.groupby('Artist')['Likes'].sum().reset_index()

# Sort the data by number of likes and select the top 10 artists
top10likes = grouped5.sort_values('Likes', ascending=False).head(10)

mpl.pyplot.figure(figsize=(12, 8))
sb.barplot(data=top10likes, x='Artist', y='Likes')
mpl.pyplot.title('Number of Likes per Artist')
mpl.pyplot.xlabel('Artist')
mpl.pyplot.ylabel('Number of Likes')
mpl.pyplot.xticks(rotation=90)

mpl.pyplot.show()

# as previous data shown BTS / Blackpink has the most likes and engaged  fans in their works and most likely comments and like
# their videos but doesnt mean it is the most popular song shown earlier by streams which is Blinding Lights
```

```python
# Explore single releases by artists and identify 
# top performers based on likes and display a table for each artist with their Views and Likes as the other columns.

singles = data.loc[data['Album_type'] == 'single']

grouped6 = singles.groupby('Artist').agg({'Views': 'sum', 'Likes': 'sum'})

topSingles = grouped6.sort_values(['Likes'], ascending=False).head(10)

for artistName in topSingles.index:
    artistSingles = data[data['Artist'] == artistName]
    table = artistSingles[['Title', 'Views', 'Likes']].sort_values(by='Likes', ascending=False)
    print(f'Top Single Songs by {artistName}:')
    display(table)
    
# It is interesting to note that the number of likes for a single release does not necessarily 
# correspond directly to the number of views it receives. Some artists may have a higher ratio of 
# likes to views than others, indicating a particularly dedicated fan base or a higher level of engagement with their audience.    
```

```python
# Exploring compilation releases by artists and identify top performers based on views and likes and display a table for 
# each artist with their Views and Likes as the other columns. Make some remark about the table.

compilation = data.loc[data['Album_type'] == 'compilation']

grouped7 = compilation.groupby('Artist').agg({'Views': 'sum', 'Likes': 'sum'})

topComp = grouped7.sort_values(['Likes'], ascending=False).head(10)

for artistName in topComp.index:
    artistComp = data[data['Artist'] == artistName]
    table = artistComp[['Title', 'Views', 'Likes']].sort_values(by='Likes', ascending=False)
    print(f'Top Compilation Songs by {artistName}:')
    display(table)
    
# table allows us to compare the performance of different compilation albums for each artist and see which ones have 
# resonated the most with their audience. We can also use this information to identify patterns or trends in 
# the types of compilation albums that are most successful for different artists.
```

```python
#Identify the top 10 channels based on views and display a bar chart of it and make some remarks.

channels = data.groupby('Channel')['Views'].sum().sort_values(ascending=False).head(10)

mpl.pyplot.figure(figsize=(12, 8))
mpl.pyplot.barh(channels.index, channels.values)
mpl.pyplot.gca().invert_yaxis()
mpl.pyplot.title('Top 10 Channels based on Views')
mpl.pyplot.xlabel('Views')
mpl.pyplot.ylabel('Channel Name')
mpl.pyplot.show()

# T-series is the most viewed channels and its mainly because of its target demographic and country
# the rest of the channels are all music oriented with cocomelon as baby channel that is famous for kids to use
```
